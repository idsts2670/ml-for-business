```python
# Data Wrangling

# disable warnings
import warnings
warnings.filterwarnings('ignore')

# library import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# matplotlib jpn support
import japanize_matplotlib

# dataframe display function
from IPython.display import display

# diaply adjustment
# numpy float display adjustment
np.set_printoptions(suppress = True, precision = 4)
# pandas float display adjustment
pd.options.display.float_format = '{:.4f}'.format
# show all columns of dataframe
pd.set_option('display.max_columns', None)
# default font
plt.rcParams['font.size'] = 14
random_seed = 123
```

```python
# import data
# breast cancer data from sklearn.datasets
from sklearn.datasets import load_breast_cancer

# load the data
cancer = load_breast_cancer()

# read annotation
print(cancer.DESCR)
```

```python
# dataframe columns change
columns = [
    '半径_平均', 'きめ_平均', '周長_平均', '面積_平均', 
    '平滑度_平均','コンパクト度_平均', '凹面_平均',
    '凹点_平均', '対称性_平均', 'フラクタル度_平均',
    '半径_標準誤差', 'きめ_標準誤差', '周長_標準誤差',
    '面積_標準誤差', '平滑度_標準誤差',
    'コンパクト度_標準誤差', '凹面_標準誤差', '凹点_標準誤差',
    '対称性_標準誤差', 'フラクタル度_標準誤差',
    '半径_最大', 'きめ_最大', '周長_最大', '面積_最大', 
    '平滑度_最大','コンパクト度_最大', '凹面_最大', '凹点_最大', 
    '対称性_最大', 'フラクタル度_最大'
]

# change loaded data to df
df = pd.DataFrame(cancer.data, columns = columns)

# get answer label
y = pd.Series(cancer.target)
```

```python
# check the data
# display from index number 20~24
display(df[20:25])
```

```python
# check the answer labels
print(y[20:25])
```

```python
# check the statistical info

# number of columns and row
print(df.shape)
print()

# number of answer labels: "1", "0"
print(y.value_counts())
```

```python
# split data
# import data with label = 0 (malignant)
df0 = df[y == 0]

# import data with label = 1 (benign)
df1 = df[y == 1]

display(df0.head())
display(df1.head())
```

```python
# scatter plot
# adjust the graph size
plt.figure(figsize = (6,6))

# scatter plot of data with response variable == 0
plt.scatter(df0['半径_平均'], df0['きめ_平均'], marker = 'x', c = 'b', label = '悪性')
# scatter plot of data with response variable == 1
plt.scatter(df1['半径_平均'], df1['きめ_平均'], marker = 's', c = 'k', label = '良性')

# show grid
plt.grid()

# show labels
plt.xlabel('半径_平均')
plt.ylabel('きめ_平均')

# show legend
plt.legend()

# show plot
plt.show()
```

```python
# Data Wrangling
# cut to two columns data
input_columns = ['半径_平均', 'きめ_平均']
x = df[input_columns]
display(x.head())
```

```python
# split train data and verification data
from sklearn.model_selection import train_test_split
# shuffle the data with random digits
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.7, test_size = 0.3, random_state = random_seed)
```

```python
# see the result dimension of split
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# see the result data of split
display(x_train.head())
display(x_test.head())
display(y_train.head())
display(y_test.head())
```

```python
# select the algorithm
from sklearn.linear_model import LogisticRegression
# reset the algorithm
algorithm = LogisticRegression(random_state = random_seed)
```

```python
# learning
algorithm.fit(x_train, y_train)
print(algorithm)
```

```python
# predict
y_pred = algorithm.predict(x_test)

# show the result
print(y_pred)
```

```python
# model evaluation
# right answer 10 labels from the beginning
y_test10 = y_test[:10].values
print(y_test10)

# predicted answer 10 from the beginning
y_pred10 = y_pred[:10]
print(y_pred10)
```

```python
# count the number of right answer

# correct answer = predicted results
w1 = (y_test10 == y_pred10)
print(w1)

# number of correct answers
w2 = w1.sum()
print(w2)
```

```python
# accuracy calculation

# number of correct answers
w = (y_test.values == y_pred)
correct = w.sum()

# number of test data
N = len(w)

# accuracy (= # of correct answers / # of test data)
score = correct / N

# show the result
print(f'精度: {score:.04f}')
```

```python
# regular ways to get accuracy
# get accuracy
score = algorithm.score(x_test, y_test)
print(f'score: {score:.04f}')
```

```python
# tuning
# improve the model from the original data
x2_train, x2_test, y_train, y_test = train_test_split(df, y, train_size = 0.7, test_size = 0.3, random_state = random_seed)

# make instance with logistic regression model again
algorithm2 = LogisticRegression(random_state = random_seed)

# learn from train data
algorithm2.fit(x2_train, y_train)

# check the accuracy by score function
score2 = algorithm2.score(x2_test, y_test)
print(f'score: {score2:.04f}')
```
